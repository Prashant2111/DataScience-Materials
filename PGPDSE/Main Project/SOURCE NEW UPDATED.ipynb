```python
n=0
```

```python
df=news_microsoft[news_microsoft.Facebook>=n]['Facebook'].values
df_scale=(df-df.min())/(df.max()-df.min())
news_microsoft.loc[news_microsoft.Facebook>=n,'Facebook']=df_scale

sourcefacebook=news_microsoft[(news_microsoft.Facebook>=n)&(news_microsoft.Facebook<10000)].groupby('Source').agg({'IDLink':'count','Facebook':['sum','min','max','mean','median']}).reset_index()
sourcefacebook.columns=['source','f_count','f_sum','f_min','f_max','f_mean','f_median']
source_df1=sourcefacebook.sort_values(by='f_mean',ascending=False)
```

```python
df=news_microsoft[news_microsoft.GooglePlus>=n]['GooglePlus'].values
df_scale=(df-df.min())/(df.max()-df.min())
news_microsoft.loc[news_microsoft.GooglePlus>=n,'GooglePlus']=df_scale

sourceGooglePlus=news_microsoft[(news_microsoft.GooglePlus>=n)&(news_microsoft.GooglePlus<10000)].groupby('Source').agg({'IDLink':'count','GooglePlus':['sum','min','max','mean','median']}).reset_index()
sourceGooglePlus.columns=['source','g_count','g_sum','g_min','g_max','g_mean','g_median']
source_df2=sourceGooglePlus.sort_values(by='g_mean',ascending=False)
```

```python
df=news_microsoft[news_microsoft.LinkedIn>=n]['LinkedIn'].values
df_scale=(df-df.min())/(df.max()-df.min())
news_microsoft.loc[news_microsoft.LinkedIn>=n,'LinkedIn']=df_scale


sourceLinkedIn=news_microsoft[(news_microsoft.LinkedIn>=n)&(news_microsoft.LinkedIn<10000)].groupby('Source').agg({'IDLink':'count','LinkedIn':['sum','min','max','mean','median']}).reset_index()
sourceLinkedIn.columns=['source','l_count','l_sum','l_min','l_max','l_mean','l_median']
source_df3=sourceLinkedIn.sort_values(by='l_mean',ascending=False)
```

```python
source_df1.merge(source_df2,on='source',how='outer').merge(source_df3,on='source',how='outer').sort_values(ascending=False,by='l_max')
```

```python
source=source_df1.merge(source_df2,on='source',how='outer').merge(source_df3,on='source',how='outer').sort_values(ascending=False,by='l_max')
```

```python
source.to_csv("source_updated.csv", index=False)
```
