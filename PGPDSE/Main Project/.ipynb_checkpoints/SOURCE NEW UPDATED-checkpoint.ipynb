```python
import pandas as pd
import numpy as np
```

```python
news=pd.read_csv('News_Data_Updated.csv')
```

```python
news.head()
```

```python
news['Topic'].value_counts()
```

```python
news=news.dropna()
```

```python
news.shape
```

```python
source=pd.DataFrame()
```

```python
source['source']=news['Source']
```

```python
source.shape
```

```python
source['IDLink']=news['IDLink']
```

New dataframe from News_final with following  columns
Source
Number of records
Sum
mean
Median
min
max

```python
source['numberofrecords']=source['IDLink'].groupby(source['source']).transform('count')
```

```python
source.groupby('source')['IDLink'].transform('count')
```

```python

source['facebook']=news['Facebook']
source['fsum']=source.facebook[source.facebook>0].groupby(source['source']).transform('sum')

source['fmean']=source.facebook[source.facebook>0].groupby(source['source']).transform('mean')

source['fmedian']=source.facebook[source.facebook>0].groupby(source['source']).transform('median')

source['fmin']=source.facebook[source.facebook>0].groupby(source['source']).transform('min')

source['fmax']=source.facebook[source.facebook>0].groupby(source['source']).transform('max')

source[source.facebook>0].sort_values(by='source',ascending=False).head()
```

```python

source['GooglePlus']=news['GooglePlus']
source['gsum']=source.GooglePlus[source.GooglePlus>0].groupby(source['source']).transform('sum')

source['gmean']=source.GooglePlus[source.GooglePlus>0].groupby(source['source']).transform('mean')

source['gmedian']=source.GooglePlus[source.GooglePlus>0].groupby(source['source']).transform('median')

source['gmin']=source.GooglePlus[source.GooglePlus>0].groupby(source['source']).transform('min')

source['gmax']=source.GooglePlus[source.GooglePlus>0].groupby(source['source']).transform('max')

source[source.GooglePlus>0].sort_values(by='source',ascending=False).head()
```

```python
source['LinkedIn']=news['LinkedIn']
source['lsum']=source.LinkedIn[source.LinkedIn>0].groupby(source['source']).transform('sum')

source['lmean']=source.LinkedIn[source.LinkedIn>0].groupby(source['source']).transform('mean')

source['lmedian']=source.LinkedIn[source.LinkedIn>0].groupby(source['source']).transform('median')

source['lmin']=source.LinkedIn[source.LinkedIn>0].groupby(source['source']).transform('min')

source['lmax']=source.LinkedIn[source.LinkedIn>0].groupby(source['source']).transform('max')

source[source.LinkedIn>0].sort_values(by='source',ascending=False).head()
```

```python
source.head()
```

```python
source=source.dropna()
```

```python
source.to_csv("source_updated.csv", index=False)
```
