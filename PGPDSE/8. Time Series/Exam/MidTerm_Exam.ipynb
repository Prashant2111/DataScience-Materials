## 2.What is the purpose of using exponential smoothing methods?



* Exponential smoothing averages or exponentially weighted moving averages consist of forecast based on previous periods data with exponentially declining influence on the older observations.

* Exponential smoothing methods are used for flattening time series data. 

* Exponential smoothing methods consist of special case exponential moving with notation ETS (Error, Trend, Seasonality) where each can be none(N), additive (N), additive damped (Ad), Multiplicative (M) or multiplicative damped (Md).

* One or more parameters control how fast the weights decay.These parameters have values between 0 and 1



## 3.What are the two types of moving averages that can be used?


Two main types of moving averages:

1) Centered moving average  - calculated as the average of raw observations at, before and after time, t.

2) Trailing moving average - uses historical observations and is used on time series forecasting.




## 4.Explain seasonality in a time series. In what ways can it be identified?


* Seasonality is the presence of variations that occur at specific regular intervals, less than a year, such as weekly, monthly, or quarterly. 
* Seasonality may be caused by various factors, such as weather, vacation, and holidays and consists of periodic, repetitive, and generally regular and predictable patterns in the levels of a time series.

* Seasonality can be identified by Decomposition of the Time Series, If seasonality is present in the data, one particular season's data may be too small or too large compared to others. 

## 5. What is a random walk model? How is it related to autocorrelation?



* A **random walk** is a mathematical object, known as a stochastic or random process, that describes a path that consists of a succession of random steps on some mathematical space such as the integers. An elementary example of a random walk is the random walk on the integer number line, Z , which starts at 0 and at each step moves +1 or −1 with equal probability.


* **How is it related to autocorrelation?**


- We calculate the correlation between each observation and the observations at previous time steps. Autocorrelation plot or a correlogram plot is a plot of these correlations. 
- We expect a strong auto-correlation with the previous observation and a linear fall oﬀ from there with previous lag values.

## 7. What is stationarity? What are the necessary conditions to determine if a series is stationary?

* When the values of Time series data are not a function of time, we have a stationary time series, the series will not have predictable pattern, only a stationary series can be forecasted.

* A stationary time series is one whose statistical properties such as mean, variance, autocorrelation, are all constant over time.The Augmented Dickey Fuller Test (ADF) is a test for stationarity. The null hypothesis is that time series is non-stationary. Alternative hypothesis is that time series is stationary.



## 8. How does the acf and pacf help in fitting an ARIMA model?


Autocorrelation of a time series by lag is called the AutoCorrelation Function. The partial autocorrelation at lag n is the correlation that results after removing the effect of any correlations due to the terms at shorter lags.acf and pacf help in fitting the best fit model

#### Dicky Fuller Test on the timeseries to verify the null hypothesis that the TS is non-stationary.

Ho=TS is non-stationary

HA=TS is stationary

```python
from statsmodels.tsa.stattools import adfuller
```

```python
dftest = adfuller(df)
print('DF test statistic is %3.3f' %dftest[0])
print('DF test p-value is %1.4f' %dftest[1])
```

* Rolling mean is clearly increasing with time and this is not a stationary series. Also, p-value is greater than Alpha(alpha=0.05), therefore we fail to reject null hypothesis, therefore it is a non stationary series

```python
plt.figure(figsize=(10, 5))
plt.plot(np.log10(df).diff(periods=1))
plt.xlabel('Years')
plt.ylabel('Differenced Log (Tractor Sales)')

```

```python
# The above is a stationary data
pass_ts_log = np.log10(df)
pass_ts_log.dropna(inplace=True)

pass_ts_log_diff = pass_ts_log.diff(periods=1) # same as ts_log_diff = ts_log - ts_log.shift(periods=1)
pass_ts_log_diff.dropna(inplace=True)
```

```python
fig, axes = plt.subplots(1, 2)
fig.set_figwidth(12)
fig.set_figheight(4)
smt.graphics.plot_acf(pass_ts_log, lags=30, ax=axes[0])
smt.graphics.plot_pacf(pass_ts_log, lags=30, ax=axes[1])
plt.tight_layout()
```

Nonstationary series have an ACF that remains significant for half a dozen or more lags, rather than quickly declining to zero. we must difference such a series until it is stationary before you can identify the process

The above ACF is “decaying”, or decreasing, very slowly, and remains well above the significance range (blue band) for at least a dozen lags. This is indicative of a non-stationary series.

```python
data['date'] = data.index
train = data[data.index < '1959-01-01']
test = data[data.index >= '1959-01-01']
train_sales_ts_log = np.log10(train['Passengers'])
```

```python
best_model = sm.tsa.statespace.SARIMAX(train_sales_ts_log,
                                      order=(0, 1, 1),
                                      seasonal_order=(1, 0, 1, 12),
                                      enforce_stationarity=True)
best_results = best_model.fit()
```

```python
print(best_results.summary().tables[0])
print(best_results.summary().tables[1])
```

```python
pred_dynamic = best_results.get_prediction(start=pd.to_datetime('1959-01-01'), dynamic=True, full_results=True)
pred_dynamic_ci = pred_dynamic.conf_int()
pred99 = best_results.get_forecast(steps=24, alpha=0.1)
# Extract the predicted and true values of our time series
sales_ts_forecasted = pred_dynamic.predicted_mean
testCopy = test.copy()
testCopy['pass_ts_forecasted'] = np.power(10, pred99.predicted_mean)
```

```python
axis = train['Passengers'].plot(label='Train Sales', figsize=(10, 6))
testCopy['Passengers'].plot(ax=axis, label='Test Sales', alpha=0.7)
testCopy['pass_ts_forecasted'].plot(ax=axis, label='Forecasted Sales', alpha=0.7)
axis.set_xlabel('Years')
axis.set_ylabel('Tractor Sales')
plt.legend(loc='best')
plt.show()
plt.close()
```

```python
# Get forecast 36 steps (3 years) ahead in future
n_steps = 36
pred_uc_99 = best_results.get_forecast(steps=36, alpha=0.01) # alpha=0.01 signifies 99% confidence interval
pred_uc_95 = best_results.get_forecast(steps=36, alpha=0.05) # alpha=0.05 95% CI

# Get confidence intervals 95% & 99% of the forecasts
pred_ci_99 = pred_uc_99.conf_int()
pred_ci_95 = pred_uc_95.conf_int()
```
