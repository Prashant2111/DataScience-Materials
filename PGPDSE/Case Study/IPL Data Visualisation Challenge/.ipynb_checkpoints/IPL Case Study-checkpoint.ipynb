```python
df.dropna(axis=0,inplace=True)
```

```python
df.shape
```

```python
df.index=np.arange(686)
```

```python
df.dtypes
```

```python
def get(x):
    r=''
    for i in x:
        if i != '/':
            r=r+i
        elif i=='/':
            break;
    return r
```

```python
df['Team1_score']=df.Team1_score.map(lambda x:get(x))
```

```python
df['Team2_score']=df.Team2_score.map(lambda x:get(x))
```

```python
fig = plt.figure(figsize=(10,10))
fig.suptitle('No of matches by Venue', fontsize=20)
sns.countplot(y=df.Match_venue, orient='v',order = df['Match_venue'].value_counts().index)
plt.xticks(rotation=0)
plt.xlabel('No of Matches', fontsize=18)
plt.ylabel('Match Venue', fontsize=16)
plt.show()
```

```python
fig = plt.figure(figsize=(12,8))
fig.suptitle('Toss_decision : Feild Vs Bat', fontsize=20)
sns.countplot(x=df.Toss_decision,order = df['Toss_decision'].value_counts().index)
plt.xticks(rotation=0)
plt.xlabel('Toss Decision', fontsize=18)
plt.ylabel('Number of Times Choosen', fontsize=16)
plt.show()
```

```python
fig = plt.figure(figsize=(12,8))
fig.suptitle('No Wins for each Team', fontsize=20)
sns.countplot(y=df.Winning_team, orient='v',order = df['Winning_team'].value_counts().index)
plt.xticks(rotation=0)
plt.xlabel('No of Wins ', fontsize=18)
plt.ylabel('Team', fontsize=16)
plt.show()
fig.savefig('No of Wins for each Team.jpg')
```
