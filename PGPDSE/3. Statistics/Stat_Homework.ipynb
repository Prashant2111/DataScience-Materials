```python
import numpy             as np
import pandas            as pd
import scipy.stats       as stats
import matplotlib.pyplot as plt
```

```python
p=0.05
n=10
k = np.arange(0,11)
binomial=stats.binom.pmf(k,n,p)
print(binomial)
```

```python
print(binomial[0])
```

```python
print(binomial[1])
```

```python
binomial1=stats.binom.cdf(k,n,p)
print(binomial[2])
```

```python
print(1-binomial[2])
```

```python
plt.plot(k,binomial, 'o-')
plt.show()
```

```python
p=0.868
n=3
k=np.arange(0,4)
binomial= stats.binom.pmf(k,n,p)
print(type(binomial))
print(binomial[3])
print(binomial[0])
binomial1= stats.binom.cdf(k,n,p)
print(1-binomial1[1])

```

```python
mean=n*p
std=(mean*(1-p))**0.5
print(mean,"\n",std)
```
