```python
t = np.arange(0, 10, 0.1)
amp = np.sin(t)
plt.plot(t, amp)
plt.title("Sine Graph")
plt.grid()
plt.show()
```

```python
t = np.arange(0, 10, 0.1)
amp = np.cos(t)
plt.plot(t, amp)
plt.title("Cosine Graph")
plt.grid()
plt.show()
```

```python
t = np.arange(0, 20, 0.1)
t1 = np.arange(0, 20, 0.1)
amp = np.sin(t)
amp1 = np.cos(t1)
plt.title("Graph")
plt.plot(t, amp,label='Sine')
plt.plot(t1, amp1,label='Cosine',linestyle='--')
plt.grid()
plt.legend()
plt.show()
```

```python
data=DataFrame([200,240,300,140,600,770,250,450],index=["Infosys","TCS","HCL","Wipro","Google","FB","Amazon","Twitter"],columns=["Revenue (in $million)"])
data
```

```python
y=data.values.flatten()                           #flatten will convert the values into a single line
y
```

```python
xpos=np.arange(data.size)
plt.bar(xpos,y,label="Revenue")
plt.xticks(xpos,data.index)  # setting tick labels on X axis ,used to replace the values of xpos with the company names
plt.ylabel('Company Revenue in $million')
plt.xlabel('Companies')
plt.title('Bar Chart')
plt.legend()
plt.show()
```

```python
xpos=np.arange(data.size)
plt.barh(xpos,y,label="Revenue")
plt.yticks(xpos,data.index)  # setting tick labels on X axis ,used to replace the values of xpos with the company names
plt.ylabel('Company Revenue in $million')
plt.xlabel('Companies')
plt.title('Bar Chart')
plt.legend()
plt.show()
```

```python
plt.bar(data.index,y)         # should apply the flattened value i.e y
plt.show()
```

```python
#histogram - y axis is frequency and x axis is range
#-a diagram consisting of rectangles whose area is proportional to the frequency of a variable and whose width is equal to the class interval.
blood_s=[113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_s)   #bins:number of bars,default=10
plt.show()
```

```python
#histogram - y axis is frequency and x axis is range
blood_s=[113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_s,bins=4,rwidth=0.8)   #bins:number of bars,default=10, rwidth is for specifing the width of the bins 
plt.show()
```

```python
blood_s=[113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_s,bins=[80,100,125,150],rwidth=0.95,color='m')   # create a histogram from 80 to 100, 100 to 125, 125 to 150
plt.show()
```

```python
bs_men=[113,85,90,150,149,88,93,115,135,80,77,82,129]
bs_woman=[67,98,89,120,133,150,84,69,89,79,120,112,100]
plt.hist([bs_men,bs_woman],bins=[80,100,125,150],rwidth=0.97,color=["green","blue"],label=['men','woman'])
#plt.xlabel()
plt.ylabel("Blood Sugar Level")
plt.legend()
plt.show()
```

```python
expense_values= [1400,600,300,410,250]
expenese_labels=["HomeRent","Food","Phone/Internet Bill","Car Maintenance","other Utilities"]
plt.pie(expense_values,labels=expenese_labels,autopct='%0.1f%%',radius=1.5,explode=[0,0.1,0,0,0.2])
plt.axis('equal')    # if we want to change the radius then we should comment plt.axis
plt.title('Pie Chart')
plt.savefig('pirchart.jpg')
plt.show()
```

```python
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
tips.tail()
```

```python
tips.describe()    # works only on Numercial Data
```

```python
tips.info()    # we can use it to check null values
```

```python
tips.nunique()
```

```python
#scatter plot
sns.lmplot(x='total_bill',y='tip',data=tips,scatter=True,fit_reg=False,height=7)
plt.show()         
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,scatter=False,fit_reg=True,height=7)
plt.show()   
```

```python
#hue 
sns.lmplot(x='total_bill',y='tip',hue='sex',data=tips,scatter=True,fit_reg=False,height=7) #legend=False
plt.show()    
```

```python
#hue 
sns.lmplot(x='total_bill',y='tip',hue='smoker',col='time',row='sex',data=tips,scatter=True,fit_reg=False,height=8,palette='Oranges',markers=['o','x']) #legend=False
plt.show()    
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,palette='rainbow',size=8,jitter=False)
plt.show()    # disadvantage is overlapping of values
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,palette='rainbow',size=8,jitter=True)
plt.show()    # disadvantage is overlapping of values
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,palette='rainbow',size=7)  # height is not available
plt.show()
```

```python
sns.countplot(x='sex',hue='time',data=tips,palette='Oranges')
plt.show()
```

```python
sns.countplot(x='smoker',hue='time',data=tips,palette='Oranges')
plt.show()
```

```python
sns.countplot(x='day',hue='time',data=tips,palette='Oranges')
plt.show()
```

```python
sns.distplot(tips['total_bill'])
plt.show()                         # the values in y axis is weights, the weight multiplied by x axis area is 1
                                 # normalized curve will have bell shape
#Skewness
# left, right , center
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,size=4)
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,size=4,kind='hex')
plt.show()
```

```python
sns.boxplot(x=tips['total_bill'])
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,whis=False) # doesnot consider the min an max values, everything outside the q1 and q3 is outlier 
plt.show()
```
